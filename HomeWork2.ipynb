```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, color='green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```

## Задание1.3*

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)
ax1.set_title('График y1')
ax2.set_title('График y2')
ax3.set_title('График y3')
ax4.set_title('График y4')
ax1.set_xlim(-5, 5)
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

## Задание1.4*

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv("creditcard.csv")
data.head()
```

```python
counts = pd.value_counts(data['Class'])
```

```python
counts.plot(kind='bar')
plt.show()
```

```python
counts.plot(kind='bar', logy=True)
plt.show()
```

```python
data0 = data.loc[data['Class'] == 0,'V1']
data1 = data.loc[data['Class'] == 1,'V1']
plt.hist(data0, bins=20, alpha=.5, density=data0.count(), color='grey', label='Class 0')
plt.hist(data1, bins=20, alpha=.5, density=data1.count(), color='red', label='Class 1')
plt.legend(loc='upper left')
plt.xlabel('V1')
plt.show()
```
